```python
import random

class Deck():
    
    def __init__(self):
        self.all_cards =[]
        for suit in suits:
            for rank in ranks:
                self.all_cards.append(Card(suit, rank))
    
    def shuffle(self):
        random.shuffle(self.all_cards)

# test 2 :Check if all the cards have been shuffled.
# deck = Deck()
# print("before shuffle",deck.all_cards[:5])
  
    
# deck.shuffle()

# print("after shuffle",deck.all_cards[:5])


    def deal_one(self):
        return self.all_cards.pop()
    
# test 3:deal one card and display it
# deck = Deck()
# deck.shuffle()

# dealt_card = deck.deal_one()
# print("Dealt card:", dealt_card)

# # test 4:show the remaining numbers of cards
# print("remining cards in deck", len(deck.all_cards) )
            
```

```python
class Hand():
    def __init__(self):
        self.cards = []
        self.value = 0
        self.aces = 0
        
    def add_card(self, card):
        self.cards.append(card)
        self.value += values[card.rank]
        if card.rank == "Ace":
            self.aces += 1
            
    def adjust_for_ace(self):
        if self.aces > 0 and self.value > 21:
            self.value -=10
            self.aces -= 1
        
```

```python
class Chips():
    def __init__(self, total):
        self.total = total
        self.bet = 0
    
    def win_bet(self):
        self.total += self.bet
        
    def lose_bet(self):
        self.total -= self.bet
        
# test 5:
# chips = Chips()
# chips.bet = 10
# chips.win_bet()
# print(f"after winning, total chips: {chips.total}")
```

```python
def take_bet(chips):
    while True:
        try:
            bet = int(input("enter 1-10 to begin bet"))
            if bet > chips.total:
                print("your bet number is over the total")
            elif bet <=  chips.total:
                return bet
    
        except ValueError:
            print("you enter a invalid value. please enter a number")

# test 6:
# chips = Chips(total = int(input("enter 1-10 to begin bet")))
# bet = take_bet(chips)
# print("final bet:", bet)
```

```python
def heat_or_stand(deck, hand):
    global playing
    
    card = deck.deal_one()
    hand.add_card(card)
    
    if hand.value > 21:
        hand.adjust_for_ace()
        
    while playing == True:
    choice = input("type 'h' or 's' to hit or stand")
    if choice == "h":
        card = deck.deal_one()
        hand.add_card(card)
        
    elif choice =="s":
        playing = False
        break
    
```

```python
def show_some(player, dealer):
    print("Player's Cards:")
    for card in player.cards:
        print(str(card))
        
    print("\nDealer's Card:")
    print(str(dealer.cards[0])) 
    print("<hidden card>")       
        
def show_all(player, dealer):
    print("Player's Cards:")
    for card in player.cards:
        print(str(card))
    print(f"Player's Hand = {player.value}")
    
    print("\nDealer's Cards:")
    for card in dealer.cards:
        print(str(card))         
    print(f"Dealer's Hand = {dealer.value}")
```
