```python
fin_data['date'] = pd.to_datetime(fin_data['date'], format = '%Y-%m-%d')
print(fin_data)
```

NaN values

```python
# 計算每個columns的空值數量
fin_data.isna().sum()
```

```python
# 從fin_data中選擇所有rows以及前20個columns
fin_data.iloc[:, :20].head()
```

```python
# ffill用每列中前一個有效值填補所有缺失值 (NaN)
# inplace=True表示改動後即覆蓋fin_data的原始數據
fin_data.ffill(inplace = True)
```

```python
fin_data.iloc[:, :20].head()
```

```python
# 處理遺漏值（NaN），刪除含有遺漏值的行，並重置索引

# missing_values列出每個欄位中缺失值的數量。
# 印出保留缺失值數量不為0的欄位
# rows_to_delete 儲存的是刪除行數的最大可能值(即最壞情況)
# 印出刪除的行數，方便檢查

missing_values = fin_data.isna().sum()
print(missing_values[missing_values != 0])
rows_to_delete = missing_values[missing_values != 0].max()
print(f'We will loose {rows_to_delete} rows')
```

```python
# 將fin_data的某個欄位(date)設為索引(index)，並直接修改原始的fin_data，不返回新 DataFrame
fin_data.set_index('date', inplace = True)
fin_data.head()
```
