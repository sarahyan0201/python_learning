3.插值法（Interpolation）
是一種填補遺漏值的方式，利用「遺漏值的前後數據」來估算出一個合理的中間值。

插值法的假設是：「遺漏的數值應該與相鄰數值之間有一定的關係，通常位於它們之間的某個位置。」
因此較適合用在連續性/ 時間序列性的資料。

(1)連續性資料：像是一段河流的水位數據，水位是平滑變化的，遺漏的水位值可以用上下游的水位推算。
(2)時間序列性資料：像每天的氣溫，今天的氣溫遺漏了，可以用昨天和明天的氣溫推算出一個合理的值。

首先先判斷哪些column中有NA值

```python
# 建立一個空的列表na_col，用來儲存有遺漏值的欄位名稱
# 使用迴圈遍歷DataFrame(就是df)內所有columns名稱
# 選取DataFrame中i那一欄，檢查該欄位中每個值是否為空值
# 若該欄位有遺漏值if條件為True，則將欄位名稱i添加到na_col列表中。

na_col = []
for i in df.columns:
    if df[i].isnull().values.any():
        na_col.append(i)
na_col
```

```python
# 建立一個df的副本，將裡面的數據複製到新變數df_fillin中
# 使用for迴圈逐一遍歷列複製的na_col中的欄位名稱
# 將用線性做為標準的插值結果存回df_fillin中對應的欄位
# 查看數據

df_fillin = df.copy()
for i in na_col:
  df_fillin[i] = df[i].interpolate(method = 'linear')
```

4. 用K-Nearest Neighbours填補遺漏值

KNN考慮數據之間的相似性，挑選想要的數據將平均值計算後結果填進空值，比簡單的均值或中位數填補更準確

```python
x = df.drop(["TypeOfSteel_A300","TypeOfSteel_A400","Fault_class"],axis=1)
```

```python
# sklearn.impute模組中導入 KNNImputer
# 初始化KNNImputer物件，設定對於每個遺漏值，找到與它「最相似」3行數據的參數，存在imputer中
# 讓imputer中的fit_transform分析數據 X，找到每一行數據之間的「相似性」，並存在df_fillk當中

from sklearn.impute import KNNImputer
imputer = KNNImputer(n_neighbors = 3)
df_fillk = imputer.fit_transform(x)
```

```python
# 從其他工具(例如 NumPy)處理過數據，回傳的結果可能沒有列名，為了保持一致性，需要將原來的列名加回來
df_fillk = pd.DataFrame(df_fillk, columns = x.columns)
df_fillk.head(10)
```

5. 用MICE填補遺漏值

【MICE 的運行流程可以分為以下幾步：】

·初始化缺失值
首先，為數據中的缺失值分配初始值，這些值通常可以用均值、中位數或其他簡單的統計量來填補。

·按列迭代填補：逐列處理缺失值
選擇一個有缺失值的欄位作為目標變數（Target）。
使用其他列作為特徵變數（Features），通過迴歸模型來預測目標列的缺失值。
用預測出的值更新目標列的缺失值。

·反覆多次迭代
依次重複對其他列進行上述步驟，直到所有缺失值都被填補。
整個過程會執行多次迭代，以提高填補的準確性，直到結果收斂。
生成多個插補數據集

MICE 通常不僅生成一組填補結果，而是生成多組數據集，每組的填補值略有不同，考慮了不確定性。

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
imp = IterativeImputer(max_iter = 100, random_state = 0 )
imp.fit(x)
```

```python
# imp.transform(x)填補數據中的缺失值，生成填補後的數據(NumPy陣列)轉換為表格型數據：
# 第二步把填補後的數據從NumPy陣列轉換成Pandas的DataFrame，並加上原始的列名稱，讓結果更直觀、更方便操作

df_fmice = imp.transform(x)
df_fmice = pd.DataFrame(df_fmice, columns = x.columns)
```
