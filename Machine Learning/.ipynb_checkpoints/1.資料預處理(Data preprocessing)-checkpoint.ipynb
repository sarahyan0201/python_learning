2.2對Columns有條件的合併(Dummy variable)

這筆資料集已經我們的類別變數轉換成Dummy variable並合併成一個欄位。
在迴歸分析（linear、logistic…等）中，當自變數為類別變數時，都要先進行轉換虛擬變數(Dummy variable)的動作，以人工變數量化類別變數，通常取值為0或1。為了處理方便，此處先將處理完的Dummy variable合併成一個欄位

```python
# 雙重方括號 [[]]是pandas中用於選取多個columns的標準方法
df[['Pastry', 'Z_Scratch', 'K_Scatch', 'Stains','Dirtiness', 'Bumps', 'Other_Faults']]
```

```python
class_name = ['Pastry', 'Z_Scratch', 'K_Scatch', 'Stains','Dirtiness', 'Bumps', 'Other_Faults']

def class_merge(x, var1, var2, var3, var4, var5, var6, var7, var8):

    if (x[var1]==1) :
        x[var8] = class_name[0]
    elif(x[var2]==1 ):
        x[var8] = class_name[1]
    elif(x[var3]==1 ):
        x[var8] = class_name[2]
    elif(x[var4]==1 ):
        x[var8] = class_name[3]
    elif(x[var5]==1 ):
        x[var8] = class_name[4]
    elif(x[var6]==1 ):
        x[var8] = class_name[5]
    elif(x[var7]==1 ):
        x[var8] = class_name[6]
        
    return x

new_df = df.apply(lambda x: class_merge(x, 'Pastry', 'Z_Scratch','K_Scatch','Stains','Dirtiness','Bumps','Other_Faults','Fault_class'), axis=1)
new_df.drop(['Pastry', 'Z_Scratch','K_Scatch','Stains','Dirtiness','Bumps','Other_Faults'], axis=1 , inplace=True)
```

```python
TypeOfSteel = ['TypeOfSteel_A300', 'TypeOfSteel_A400']

def TypeOfSteel_merge(x, var1, var2, var3):

    if (x[var1]==1) :
        x[var3] = TypeOfSteel[0]
    elif(x[var2]==1 ):
        x[var3] = TypeOfSteel[1]
        
    return x

new_df = new_df.apply(lambda x: TypeOfSteel_merge(x, 'TypeOfSteel_A300', 'TypeOfSteel_A400', 'TypeOfSteel'), axis=1)
new_df.drop(['TypeOfSteel_A300', 'TypeOfSteel_A400'], axis=1 , inplace=True)
```

```python
new_df.head()
```

```python
# 也可以利用同樣的概念創造Dummy variable或是不同複雜的變數運算及處理(分群)
def TypeOfSteel_Dummy(x, var1, var2, var3):

    if (x[var1]== 'TypeOfSteel_A300') :
        x[var2] = 1
        x[var3] = 0
        
    elif(x[var1]=='TypeOfSteel_A400' ):
        x[var2] = 0
        x[var3] = 1
        
    return x

df = new_df.apply(lambda x: TypeOfSteel_Dummy(x, 'TypeOfSteel', 'TypeOfSteel_A300', 'TypeOfSteel_A400'), axis=1)
df.drop(['TypeOfSteel'], axis=1 , inplace=True)
df.head()
```

3. 離群值偵測 Outlier Detection
離群值代表對比其他資料下，某些資料的值與大部分資料產生極大差距，對後續分析及建模可能產生影響，需要注意資料的分配情形以利判斷。

常見離群值的出現有幾種
(1)由於資料輸入錯誤產生特殊值。
(2)特殊值可能代表特殊事件或重要的業務情境。例如機器故障、個人消費習慣，需要多方判斷。

以下利用兩種基本的方式介紹如何偵測離群值，本資料不太服從常態分配，導致 outlier 的狀況較多，如要保留這些被判斷是 outlier 的資料可以先進行資料轉換(取 log /開根號等等方式)讓資料較接近常態分配。

3.1 以標準化分數判斷
將資料轉成標準化分數或Z分數進行判斷，根據常態分配特性：

約 68% 的數據在平均值 ± 1 標準差內。
約 95% 的數據在平均值 ± 2 標準差內。
約 99.7% 的數據在平均值 ± 3 標準差內。

判斷離群值：Z 分數超過 ±3（或其他定義的閾值）的數據通常被視為離群值。

```python
from sklearn.preprocessing import StandardScaler
x = df.drop(["TypeOfSteel_A300", "TypeOfSteel_A400"],axis = 1)

# 移除非數值型資料
x = df.drop(["TypeOfSteel_A300", "TypeOfSteel_A400"], axis=1)
x = x.select_dtypes(include=['float64', 'int64'])  # 保留數值型

# 標準化數據
ss = StandardScaler()
scaled_df = ss.fit_transform(x)
scaled_df = pd.DataFrame(scaled_df, columns=x.columns)

print(scaled_df)
```

```python
# 使用drop()，從原始DataFrame刪除包含離群值的行，並將結果賦值給df_no_outlier。
df_no_outliar = df.drop(outliar_index.index)
print(df_no_outliar.shape[0])
```

```python
numeric_df = df.select_dtypes(include=['number'])
q1 = numeric_df.quantile(0.25)
q3 = numeric_df.quantile(0.75)
iqr = q3 - q1
```

```python
print(iqr)
```

```python
# 定義離群值標的
def detect_outliers(df):
    q1 = df.quantile(0.25)
    q3 = df.quantile(0.75)
    iqr = q3 - q1
    fence_low = q1 - 1.5 * iqr
    fence_high = q3 + 1.5 * iqr
    outleirs = pd.DataFrame()
    for index, value in enumerate(df.columns):
        outleirs[value] = (df.iloc[:, index] < fence_low[index]) | (df.iloc[:, index] > fence_high[index])
    return outleirs
```

```python
# 篩選要刪除的數值型列
x = df.drop(["TypeOfSteel_A300", "TypeOfSteel_A400", "Fault_class"], axis=1).select_dtypes(include='number')

# 檢測離群值
outlier_df = detect_outliers(x)
outlier_df.head()
```
